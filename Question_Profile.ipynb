```python
#Import Libraries
import xml.etree.ElementTree
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.tokenize import RegexpTokenizer 
from collections import Counter

#Get root tag of XML
e = xml.etree.ElementTree.parse('Posts.xml').getroot()
stop_words = set(stopwords.words('english'))
tokenizer = RegexpTokenizer(r'\w+')
```

```python

BoW={}

#Get each question
for atype in e.findall('row'):
    #Get Data
    body=atype.get('Body')
    que_id=int(atype.get('Id'))
    #clean
    body=body.replace("<p>","")
    body=body.replace("</p>","")
    
    #Tokenization
    body_words=tokenizer.tokenize(body)
    
    #StopWord Removal
    body_words = [word.lower() for word in body_words if not word in stop_words]
    body_words=dict(Counter(body_words))
    BoW[que_id]=body_words
print(BoW)
```
